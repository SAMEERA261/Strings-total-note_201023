### Formatting string ():
* F-strings allow you to directly embed variables and expressions inside curly braces within a string.

* commonly used methods are %-formatting, str.format(), and f-strings:

1. %-formatting:
    This is an older method that uses the % operator. You create a string with placeholders, and then use the % operator to         replace them with values.
2. str.format():
    This method uses the format() method to substitute values into placeholders within a string. 
3. f-strings (formatted string literals):
    Introduced in Python 3.6, f-strings provide a concise and readable way to embed expressions inside string literals.
   

```python
#example for raw string 

print('c:\abc:\jps')

print(r'c:/abc/newfolder/j.jpg')

print(r'c://sameera//new//old/j.jpg')
```

```python
# example for formatting string

name = "Sameera" # %-formatting
age = 23

formatted_string = "My name is %s and I am %d years old." % (name, age)
print(formatted_string)

name = "Sameera" #str.format()
age = 23

formatted_string = "My name is {} and I am {} years old.".format(name, age)
print(formatted_string)

name = "sameera" #f-strings (formatted string literals)
age = 23

formatted_string = f"My name is {name} and I am {age} years old."
print(formatted_string)


print(f'a sphere with diameter {10} cms,volume will be {(22/7)*(10/2)**3*(4/3)} cubic cms')#formatting string example
```

## In Python Strings are immutable

* </b> Strings are <u>immutable </u> in programming languages like Python  for <i>efficiency, hashing, thread safety, and maintaining consistency</i>. Immutability ensures that when you manipulate a string, a new one is created instead of modifying the existing one, making operations more memory-efficient, especially with large strings. Immutability also enables hash stability, crucial for data structures like dictionaries, and ensures thread safety, as concurrent modifications can't lead to unexpected behavior. The consistency and predictability offered by immutability simplify code understanding and maintenance, providing a reliable foundation for string manipulation in various contexts.

* A string is immutable because we can't change its value after defining it. Furthermore, once we assign a value to a string, it receives an id based on that value we assigned. Hence if we change the value of that string, it causes to change its id. Therefore, strings are immutable.
In Python, strings are immutable. That means the characters of a string cannot be changed. For example,

```python
message = 'Hola Amigos'

message[0] = 'H'

print(message)
```

* However, we can assign the variable name to a new string. For example,

```python
message = 'Hola Amigos'

# assign new string to message variable
message = 'Hello Friends'

prints(message); # prints "Hello Friends"
```
